```python
#create a custom function
import numpy as np

def custom_metric(y, y_pred):
    tp = np.where((y == 1) & (y_pred == 1), (100), 0)
    fp = np.where((y == 0) & (y_pred == 1), -5, 0)
    return np.sum([tp,fp])

# add metrics to pycaret outside of the function definition
add_metric('custom_metric', 'Custom Metric', custom_metric)
```

```python
#compairing models again
compare_models()
```

```python
#remove custom metrics
remove_metric('custom_metric')
```
