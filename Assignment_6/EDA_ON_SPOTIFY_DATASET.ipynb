```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
df1=df[["valence","energy"]]
```

```python
df1
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
#predict the labels of clusters.
label = kmeans.fit_predict(df1)
```

```python
print(label)
```

```python
#filter rows of original data
filtered_label0 = df1[label == 0]
filtered_label1 = df1[label == 1]
#plotting the results
plt.scatter(filtered_label0["valence"] , filtered_label0["energy"] , color = "red")
plt.scatter(filtered_label1["valence"] , filtered_label1["energy"] , color = "green")
plt.show()
```

```python
filtered_label1 = df1[label == 1]
filtered_label1
```

```python
print("cluster 0 Range of valence : ",filtered_label0['valence'].min(),":",filtered_label0['valence'].max())
print("cluster 0 Range of energy : ",filtered_label0['energy'].min(),":",filtered_label0['energy'].max())

print("\n\ncluster 1 Range of valence : ",filtered_label1['valence'].min(),":",filtered_label1['valence'].max())
print("cluster 1 Range of energy : ",filtered_label1['energy'].min(),":",filtered_label1['energy'].max())
```
