```python
import pandas as pd

# Load the user's uploaded dataset to take an initial look
file_path = '/content/amazon.csv'
df = pd.read_csv(file_path)

# Display basic information about the dataset
df_info = df.info()
df_info
```

```python
df_head = df.head()
df_head
```

```python
# Clean and convert columns to appropriate data types

# Remove currency symbols and convert to numeric for 'discounted_price' and 'actual_price'
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')

# Remove percentage symbol and convert to numeric for 'discount_percentage'
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%', ''), errors='coerce')

# Convert 'rating' and 'rating_count' to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', ''), errors='coerce')

# Check for null values and ensure conversions were successful
cleaned_info = df.info()
df.describe(include='all')
```

```python
from IPython.display import HTML

html_code_pie_chart = '''
<!DOCTYPE html>
<meta charset="utf-8">
<style>
  .arc path {
    stroke: #ffffff;
  }

  .legend {
    font-size: 12px;
  }
</style>
<div id="chart"></div>
<script src="https://d3js.org/d3.v5.min.js"></script>
<script>
  var data = [
    {"category": "USB Cables", "count": 233},
    {"category": "Smart Watches", "count": 76},
    {"category": "Smartphones", "count": 68},
    {"category": "Smart TVs", "count": 63},
    {"category": "In-Ear Headphones", "count": 52},
    {"category": "Remote Controls", "count": 49},
    {"category": "Mixer Grinders", "count": 27},
    {"category": "HDMI Cables", "count": 24},
    {"category": "Dry Irons", "count": 24},
    {"category": "Mice", "count": 24}
  ];

  var width = 500, height = 500, radius = Math.min(width, height) / 2;

  var color = d3.scaleOrdinal(d3.schemeCategory10);

  var svg = d3.select("#chart").append("svg")
      .attr("width", width)
      .attr("height", height)
    .append("g")
      .attr("transform", "translate(" + width / 2 + "," + height / 2 + ")");

  var arc = d3.arc()
      .outerRadius(radius - 10)
      .innerRadius(0);

  var labelArc = d3.arc()
      .outerRadius(radius - 40)
      .innerRadius(radius - 40);

  var pie = d3.pie()
      .sort(null)
      .value(function(d) { return d.count; });

  var g = svg.selectAll(".arc")
      .data(pie(data))
    .enter().append("g")
      .attr("class", "arc");

  g.append("path")
      .attr("d", arc)
      .style("fill", function(d) { return color(d.data.category); });

  g.append("text")
      .attr("transform", function(d) { return "translate(" + labelArc.centroid(d) + ")"; })
      .attr("dy", "0.35em")
      .text(function(d) { return d.data.category; });

  // Legend
  var legend = svg.append("g")
      .attr("transform", `translate(${-(width / 2) + 10}, ${-(height / 2) + 10})`);

  legend.selectAll("rect")
      .data(data)
    .enter().append("rect")
      .attr("x", 0)
      .attr("y", (d, i) => i * 20)
      .attr("width", 10)
      .attr("height", 10)
      .style("fill", d => color(d.category));

  legend.selectAll("text")
      .data(data)
    .enter().append("text")
      .attr("x", 20)
      .attr("y", (d, i) => i * 20 + 9)
      .text(d => d.category)
      .attr("class", "legend");
</script>
'''

HTML(html_code_pie_chart)
```
